```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
import yfinance as yf
```

```python
data = pd.read_csv('台股大盤成分股.csv')
data['Ticker'] = data['Ticker'].apply(str)
data
```

```python
close_df = pd.DataFrame()
open_df = pd.DataFrame()
ticker = data['Ticker'].tolist()
for i in ticker:
    stock = yf.download(i+'.TW', start='2020-01-01', end='2023-11-30')
    stock_open = stock[['Open']]
    stock_close = stock[['Close']]
    stock_open.rename(columns={'Open': i[:4]}, inplace=True)
    stock_close.rename(columns={'Close': i[:4]}, inplace=True)
    open_df = pd.concat([open_df, stock_open], axis=1)
    close_df = pd.concat([close_df, stock_close], axis=1)
```

```python
open_df.to_csv('N_days_overreaction_open_price.csv')
close_df.to_csv('N_days_overreaction_close_price.csv')
```

```python
pct_change = close_df.pct_change()
pct_change.dropna(axis=0, how='all', inplace=True)
pct_change
```

```python
training_data = pct_change.iloc[:int(len(pct_change)*0.8)]
testing_data = pct_change.iloc[int(len(pct_change)*0.8):]
```

```python
training_data
```

# Training

```python
top_n = 5
best_n = 0
best_return = 0
best_return_list = []

for N in range(30):
    N_days_rolling_pct_change = (1+training_data).rolling(window=N).apply(np.prod, raw=True) - 1
    N_days_rolling_pct_change.dropna(axis=0, how='all', inplace=True)

    daily_return_list = []
    cum_return = 0
    # exclude the last date
    for i in range(len(N_days_rolling_pct_change)-1):
        daily_top = N_days_rolling_pct_change.iloc[i].nlargest(top_n).index.tolist()
        daily_bottom = N_days_rolling_pct_change.iloc[i].nsmallest(top_n).index.tolist()

        long_openPosition = open_df[open_df.index == N_days_rolling_pct_change.index[i+1]][daily_top].sum(axis=1).values[0]
        short_openPosition = open_df[open_df.index == N_days_rolling_pct_change.index[i+1]][daily_bottom].sum(axis=1).values[0]

        long_closePosition = close_df[close_df.index == N_days_rolling_pct_change.index[i+1]][daily_top].sum(axis=1).values[0]
        short_closePosition = close_df[close_df.index == N_days_rolling_pct_change.index[i+1]][daily_bottom].sum(axis=1).values[0]

        daily_return = (long_closePosition-long_openPosition) + (short_openPosition-short_closePosition)
        cum_return += daily_return

        daily_return_list.append(daily_return)
        
    if cum_return > best_return:
        best_return = cum_return
        best_n = N
        best_return_list = daily_return_list
    
```

```python
print('Best N:', best_n)
print('Best Return:', best_return)
print('Best Return List:', best_return_list)
```
